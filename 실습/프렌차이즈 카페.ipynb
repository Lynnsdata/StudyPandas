### 부산 수영구, 서구의 프렌차이즈 카페 입점 비율

### base

```python
import pandas as pd
```

```python
df1 = pd.read_csv('부산광역시 서구_개인카페_20211116.csv',encoding='cp949')
```

```python
df2 = pd.read_csv('부산광역시 서구_프랜차이즈카페_20211116.csv',encoding='cp949')
```

```python
df3 = pd.read_csv('부산광역시 수영구_개인 카페 및 프랜차이즈 카페 현황_20211111.csv',encoding='cp949')
```

```python
data_s = pd.concat([df1,df2],ignore_index=True)
```

```python
data_f = data_s.rename(columns={'업종':'업종명'})
```

```python
result = pd.concat([data_f,df3],ignore_index=True)
```

```python
del result['소재지전화']
```

```python
result
```

```python
cafe = result.fillna('2021-11-11')
```

```python
cafe
```

```python
cafe.loc[cafe['소재지(도로명)'].str.contains('수영구'),'소재지(도로명)'] = '수영구'
cafe.loc[cafe['소재지(도로명)'].str.contains('서구'),'소재지(도로명)'] = '서구'
cafe
```

```python
cafe_list = cafe.rename(columns={'소재지(도로명)':'소재지'})
```

```python
cafe_list
```

### suyeonggu

```python
su = cafe_list[cafe_list['소재지'] == '수영구']
su.index = range(len(su))
print('수영구 내 카페 점포 수 : ', len(su))
su
```

```python
star = su[su['업소명'].str.contains('스타벅스')]
star.index = range(len(star))
print('수영구 내 스타벅스 점포 수 : ', len(star))
star
```

```python
ed = su[su['업소명'].str.contains('이디야')]
ed.index = range(len(ed))
print('수영구 내 이디야 점포 수 : ', len(ed))
ed
```

```python
cp = su[su['업소명'].str.contains('컴포즈')]
cp.index = range(len(cp))
print('수영구 내 컴포즈 점포 수 : ', len(cp))
cp
```

```python
c051 = su[su['업소명'].str.contains('카페051')]
c051.index = range(len(c051))
print('수영구 내 카페051 점포 수 : ', len(c051))
c051
```

```python
ten = su[su['업소명'].str.contains('텐퍼센트')]
ten.index = range(len(ten))
print('수영구 내 텐퍼센트 점포 수 : ', len(ten))
ten
```

```python
print('\t<< 수영구 커피전문점 중 프렌차이즈 입점 비율 >>')
print()
print('\t프렌차이즈 입점 비율 : %.3f%%'\
      %((len(star) + len(ed) + len(cp) + len(c051) + len(ten))/ len(su)*100))
print('\t1. 컴포즈 : %.3f%%' % (len(cp)/ len(su)*100))
print('\t2. 이디야 : %.3f%%' % (len(ed)/ len(su)*100))
print('\t3. 스타벅스 : %.3f%%' % (len(star)/ len(su)*100))
print('\t4. 카페051 : %.3f%%' % (len(c051)/ len(su)*100))
print('\t5. 탠퍼센트 : %.3f%%' % (len(ten)/ len(su)*100))
```

### seogu

```python
seo = cafe_list[cafe_list['소재지'] == '서구']
seo.index = range(len(seo))
print('서구 내 카페 점포 수 : ', len(seo))
seo
```

```python
star1 = seo[seo['업소명'].str.contains('스타벅스')]
star1.index = range(len(star1))
print('서구 내 스타벅스 점포 수 : ', len(star1))
star1
```

```python
ed1 = seo[seo['업소명'].str.contains('이디야')]
ed1.index = range(len(ed1))
print('서구 내 이디야 점포 수 : ', len(ed1))
ed1
```

```python
cp1 = seo[seo['업소명'].str.contains('컴포즈')]
cp1.index = range(len(cp1))
print('서구 내 컴포즈 점포 수 : ', len(cp1))
cp1
```

```python
c0511 = seo[seo['업소명'].str.contains('카페051')]
c0511.index = range(len(c0511))
print('서구 내 카페051 점포 수 : ', len(c0511))
c0511
```

```python
ten1 = seo[seo['업소명'].str.contains('텐퍼센트')]
ten1.index = range(len(ten1))
print('서구 내 텐퍼센트 점포 수 : ', len(ten1))
ten1
```

```python
print('\t<< 서구 커피전문점 중 프렌차이즈 입점 비율 >>')
print()
print('\t프렌차이즈 입점 비율 : %.3f%%'\
      %((len(star1) + len(ed1) + len(cp1) + len(c0511) + len(ten1))/ len(seo)*100))
print('\t1. 컴포즈 : %.3f%%' % (len(cp1)/ len(seo)*100))
print('\t2. 이디야 : %.3f%%' % (len(ed1)/ len(seo)*100))
print('\t3. 탠퍼센트 : %.3f%%' % (len(ten1)/ len(seo)*100))
print('\t4. 스타벅스 : %.3f%%' % (len(star1)/ len(seo)*100))
print('\t5. 카페051 : %.3f%%' % (len(c0511)/ len(seo)*100))
```

### graph

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
list_su = [len(cp)/ len(su)*100,len(ed)/ len(su)*100,len(star)/ len(su)*100,len(c051)/ len(su)*100,len(ten)/ len(su)*100]
list_su
```

```python
list_seo = [len(cp1)/ len(seo)*100,len(ed1)/ len(seo)*100,len(ten1)/ len(seo)*100,len(star1)/ len(seo)*100,len(c0511)/ len(seo)*100]
list_seo
```

```python
plt.rcParams['font.family']='NanumBarunpen'
plt.rcParams['font.size']=15
```

```python
labels = ['컴포즈','스타벅스','이디야','텐퍼센트','카페051']
sy_cafe = [len(cp)/ len(su)*100,len(ed)/ len(su)*100,len(star)/ len(su)*100,len(c051)/ len(su)*100,len(ten)/ len(su)*100]
sg_cafe = [len(cp1)/ len(seo)*100,len(ed1)/ len(seo)*100,len(ten1)/ len(seo)*100,len(star1)/ len(seo)*100,len(c0511)/ len(seo)*100]


x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots(figsize=(10,8))
rects1 = ax.bar(x - width/2, sy_cafe, width, label='Suyeong 수영구')
rects2 = ax.bar(x + width/2, sg_cafe, width, label='Seo 서구')

ax.set_ylabel('비율')
ax.set_title('프렌차이즈 입점 현황')
ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

fig.tight_layout()

plt.show() 
```
